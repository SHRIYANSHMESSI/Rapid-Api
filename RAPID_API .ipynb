```python
resp
for movie in resp['results']:
    image_url = movie.get('imageurl', ['none'])[0] if movie.get('imageurl') else 'none'
    genre = ', '.join(movie.get('genre', ['none']))
    title = movie.get('title', 'none')
    imdb_rating = movie.get('imdbrating', 'none')
    release_year = movie.get('released', 'none')
    movie_type = movie.get('type', 'none')
    synopsis = movie.get('synopsis', 'none')

    print('Title:', title)
    print('Genre:', genre)
    print('IMDB Rating:', imdb_rating)
    print('Release Year:', release_year)
    print('Movie Type:', movie_type)
    print('Synopsis:', synopsis)
    print('Image URL:', image_url)
    print()
```
